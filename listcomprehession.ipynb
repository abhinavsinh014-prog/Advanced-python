```python
lst = [1,2,3,4,5]
lsd = []

for i in lst:
    lsd.append(i**2)

print(lsd)
```

```python
lst = [1,2,3,4,5]
squared = [i**2 for i in lst]
print(squared)
```

```python
lst = [1,2,3,4,5]
env = [n**2 for n in lst if n%2==0]
print(env)
```

```python
list = [[1,2,3],[4,5,6],[7,8,9]]
flattened = [num for sublist in list for num in sublist]
print(flattened)
```

```python
word = ["hello","world","python","programming"]
lengths = [w[0] for w in word]
print(lengths)
```
