```python
class person:
    def __init__(self, name, age):
        self.__name = name
        self.__age = age
    def get_name(self):
        return self.__name
```

```python
getp = person("Abhinav", 22)
print(getp.get_name())  # This will print "Abhinav"
```

```python
class son:
    def __init__(self, name, age):
        self._name = name
        self._age = age

class sdl(son):
    def __init__(self, name, age):
        super().__init__(name, age)
    def get_name(self):
        return self._name 
     
get_sdl = sdl("Abhinav", 22)
print(get_sdl.get_name())   
```

```python
from dataclasses import dataclass

@dataclass
class WWE:
    name: str
    age: int
    fees: float


class sdl(WWE):
    def __init__(self, name, age, fees):
        super().__init__(name, age, fees)
    def get_name(self):
        return self._name
    def get_fees(self):
        return self.__fees
```
