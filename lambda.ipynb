```python
people = [
    {"name": "abhinav", "age": 24},
    {"name": "john", "age": 30},
    {'name': 'sj', 'age': 24},
    {'name': 'nam', 'age': 24},
    {"name": "doe", "age": 22}
]
```

```python
sorted(people, key=lambda x: x["age"])
```

```python
sorted(people, key=lambda x: (x["age"],x["name"]))
```

```python
from intertools import groupby

words = ["abhinav", "john", "sj", "nam", "doe"]
group = groupby(sorted(words), key=lambda x: x[0])

for key, value in group:
    print(key, list(value))
```
