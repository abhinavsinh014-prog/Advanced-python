```python
study = {
    "subject": ["Mathematics", "Electrical", "Physics","PPS","EVS"],
    "topics": [5,5,5,5,5],
    "exams DATE": ["2024-07-01","2024-07-02","2024-07-03","2024-07-04","2024-07-05"]
}

```

```python
import pandas as pd
df = pd.DataFrame(study)
print(df)
```

```python
df["preparation"] = [2,5,3,4,2]
df["number_of_topics"] = df["topics"] - df["preparation"]
print(df)
```

```python
df["preparation"] = df["preparation"].replace({2: "good",3: "excellent",0: "bad",1: "average",4: "outstanding",5: "perfect"})
print(df)
```

```python
average_preparation = df["preparation"].value_counts().get("average", 0)
print("Number of subjects with average preparation:", average_preparation)
```

```python
a = df["preparation"].value_counts()
print(a)
```
