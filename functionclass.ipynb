```python
from dataclasses import dataclass

@dataclass
class ring:
    name:str
    wins:int
    debut:int

rinfs = ring("john cena", 17, 2002)
```

```python
rinfs.name
```

```python
rinfs.wins
```

```python
@dataclass
class stats:
    name:str
    runs:int
    wickets:int=0

stat = stats("abhinav", 160)
```

```python
stat
```

```python
@dataclass(frozen=True)
class stats:
    name:str
    runs:int
    wickets:int

stat = stats("abhinav", 160, 10)
```

```python
stat.wickets = 8
```

```python
#inheritance 

@dataclass
class player:
    name:str
    position:str

@dataclass
class performance(player):
    runs:int
    wickets:int
```

```python
play = player("abhinav", "batsman")
plays = performance("Abhinav", "batsman", 160, 0)
```
